```python
def lexicographic_order(n, ordered_str_list, prefix, string):
    if len(prefix)==n:
        #string.append(prefix)
        print(prefix)
    else:
        for symbol in ordered_str_list:
            lexicographic_order(n, ordered_str_list, prefix+symbol, string)
```

```python
provided_str = "A B C D"
ordered_str_list = provided_str.split(" ")
#print(ordered_str_list)
n= 4
lexicographic_order(n, ordered_str_list,"",[])
```
